```python
# Імротуємо бібліотеки
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
import eli5
from eli5.sklearn import PermutationImportance
from sklearn.preprocessing import StandardScaler
import pickle

from plotly.graph_objects import *
from plotly.offline import init_notebook_mode,iplot
init_notebook_mode(connected=True)
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib inline
```

<h3>Завантажуємо датасет<h3>

```python
housing = fetch_california_housing()
```

```python
# Інформація про датасет 

print(housing['DESCR'])
```

```python
# Створюємо датафрейм 

df = pd.DataFrame(data = np.c_[housing['data'], housing['target']],
                 columns= housing.feature_names + ['target'])
df.head()
```

<h3>Візуалізація фічей<h3>

```python
sns.set(font_scale=1)
sns.set_style('whitegrid')

for feature in housing.feature_names:
    plt.figure(figsize=(7, 4))
    sns.scatterplot(data=df, x=feature,y='target',
                    hue='target', palette='cool', legend=False)
```

Створюємо х та у

```python
x = df.drop(columns=['target'])
print(x.shape)
    
y = df['target']
print(y.shape)
```

<h3>Підбір фічей до модельки<h3>

```python
# Подивимось на кореляцію усіх фічей в датасеті між собою

plt.figure(figsize=(8,6))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
# Рахуємо коефіцієнт корреляції усіх фічей до таргету
cor_target = abs(cor["target"])

# Подивимось на фічі з кореляцією більше 10% до таргету
relevant_features = cor_target[cor_target>0.10]
relevant_features
```

Найбільший коефіцієнт корреляції до таргету має фіча MedInc (69%). Ми бачимо, що Latitude і Longtitude коррелюють між собою на 92% по модулю, що дуже логічно і AveRooms коррелює до AveBedrms на 85%. Для побудови модельки і уникнення проблеми мультиколінеарності AveBedrms не буду використовувати.

```python
# Подивимось на дисперсію по фічах
numeric_col = list(df.select_dtypes(include=np.number).columns)

for col in numeric_col:
    print(f'{col}:  {df.var()[col]}')
```

Найменшу волатильність має AveBedrms, а отже і цікавість такої фічі для використання в модельці майже мінімальна, тому впевнено її виключаємо із навчання модельки.

```python
print(df.var()['AveBedrms'])
```

<h3>Стандартизуємо дані<h3>

```python
# Розбиваємо на тренувальну і тестову вибірку
x = x.drop(columns=['AveBedrms']) # Видаляємо фічу не має багато значущості для навчання моделі
x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle=True, test_size=0.2) 
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

<h3>Відбираємо фічі за допомогою методу permutation importance<h3>

```python
# RandomForest Classifier
rf_model_pmi = RandomForestRegressor(n_estimators=100, random_state=0).fit(x_train_scaled, y_train)
```

```python
perm = PermutationImportance(rf_model_pmi, random_state=1).fit(x_test_scaled, y_test)
eli5.show_weights(perm, feature_names = x_test.columns.tolist())
```

Отже, значення feature Permutation Importance показують нам, найбільш важливими є MedInc, Latitude, Longitude, AveOccup, так як при рандомному перемішуванні значень цих фічей, якість предікту моделькою значно погіршується. Але фічей з від'ємним скором немає, а отже нас для начання модельки задовільняють усі фічі.

<h3>Навчання моделі<h3>

```python
param_grid = {
    'max_depth': [20,50],
    'max_features' : [5,7],
    'n_estimators': [50,100,200],
    'min_samples_split': [30,50,70]}

grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=0), 
param_grid=param_grid,cv=KFold(n_splits=5, shuffle=True, random_state=1))


grid_search.fit(x_train_scaled, y_train)
```

```python
#The best score by grid search
grid_search.best_score_
```

```python
print(f'n_estimators: {grid_search.best_params_["n_estimators"]}')
print(f'max_features: {grid_search.best_params_["max_features"]}')
print(f'max_depth: {grid_search.best_params_["max_depth"]}')
print(f'min_samples_split: {grid_search.best_params_["min_samples_split"]}')
```

```python
rf = RandomForestRegressor(
                           n_estimators = grid_search.best_params_['n_estimators'],
                           max_features = grid_search.best_params_['max_features'],
                           max_depth = grid_search.best_params_['max_depth'],
                           min_samples_split=grid_search.best_params_['min_samples_split'],
                           random_state = 0)
    
rf.fit(x_train_scaled, y_train)

n_scores = cross_val_score(rf, x_train_scaled, y_train,
                           cv=KFold(n_splits=5, shuffle=True, random_state=1)).mean()

print(f'Середнє значення R^2 на 5 розбиттях по кросс валідації = {n_scores}')
```

```python
#R2 score на тестових даних
print(f'Значення R^2 на тестовій вибірці - {rf.score(x_test_scaled,y_test)}')
```

```python
# Рахуємо значенння модельки для тренувальних і тестових
y_pred_train = rf.predict(x_train_scaled)
y_pred_test = rf.predict(x_test_scaled)

# Print metrics
print(f'The R2 score train data: {metrics.r2_score(y_train, y_pred_train)}')
print(f'The MSE train data: {metrics.mean_squared_error(y_train, y_pred_train)}')
print(f'The MAE train data: {metrics.mean_absolute_error(y_train, y_pred_train)}\n')
print(f'The R2 score test data: {metrics.r2_score(y_test, y_pred_test)}')
print(f'The MSE test data: {metrics.mean_squared_error(y_test, y_pred_test)}')
print(f'The MAE test data: {metrics.mean_absolute_error(y_test, y_pred_test)}')


# Графік тестових даних предікт vs справжні значення
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred_test, label="predicted")
plt.title("California housing test and predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```

<h3>Зберігаємо модельку в Pickle File<h3>

```python
features = list(x_train.columns)
rf_model = {'features':features,
           'model': rf}
```

```python
# Зберігаємо модель в pkl файл
with open('california_housing_rf.pkl', 'wb') as file:
    pickle.dump(rf_model, file)
```

```python
# Перевіряємо, що моделька зберіглась і можна її считати
with open('california_housing_rf.pkl', mode='rb') as f:
    model_data = pickle.load(f)
```
