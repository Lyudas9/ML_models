```python
# Імротуємо бібліотеки
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV#, SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler
import math

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('../Machine-Learning-2/datasets/diabetes2.csv')
```

```python
df.head()
```

```python
#Подивимось на розподіл класів
df['Outcome'].value_counts()
```

```python
features = list(df.columns)
features.remove('Outcome')
```

```python
features
```

<h3>Намалюємо гістограми розподілу по усіх фічах в датасеті для двох класів<h3>

```python
sns.set(font_scale=1)
sns.set_style('whitegrid')

for feature in features:
    plt.figure(figsize=(5, 3))
    plt.suptitle(f'Hist by {feature}')
    df.groupby('Outcome')[feature].hist(legend=True)
```

- Одразу видно, що діабет частіше зустрічається у жінок, що були вагітні
- Глюкоза також чим більша тим більша ймовірність мати діабет
- Також у віці від 30-55 висока ймовірність мати діабет

<h3>Подивимось на кореляцію між фічами<h3>

```python
# Подивимось на корреляцію усіх фічей в датасеті між собою

plt.figure(figsize=(8,6))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
# Рахуємо коефіцієнт корреляції усіх фічей до таргету
cor_target = abs(cor["Outcome"])

# Подивимось на фічі з корреляцією більше 10% до таргету
relevant_features = cor_target[cor_target>0.10]
relevant_features
```

Бачимо що між собою коррелюють на 54% фічі Age та Pregnancies, тому для модельки залишимо тільки Age, який коррелює до нашого тагрета більше, і це майже 24%.

<h3>Feature importance<h3>

```python
features.remove('Pregnancies')
model = LogisticRegression(random_state=0).fit(df[features].values, df['Outcome'].values)

# Отримуємо скор
score = model.score(df[features].values, df['Outcome'].values)
print(score)
wo = model.intercept_[0]
w = model.coef_[0]

# Важливі фічі
feature_importance = pd.DataFrame(features,columns = ['feature'])
feature_importance['importance'] = pow(math.e, w)
feature_importance = feature_importance[:10].sort_values(by = ["importance"], ascending=False)

ax = feature_importance.sort_values(by = ["importance"], ascending=True).plot.barh(x='feature',
                                                                    y='importance')

plt.title('Important Features')
plt.show()

feature_importance
```

```python
# Створюємо Х та Y
important_features = feature_importance['feature'].to_list()

x = df[important_features]
print(x.shape)
    
y = df['Outcome']
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle=True, test_size=0.2)
```

```python
logit_model = sm.Logit(y_train, x_train)
result = logit_model.fit()
print(result.summary2())
```

Найбільш важливі фічі показують p values менше ніж 0.05. Такі фічі мають найбільш вагомий вклад в предикт таргету.
Отже найбільш вагомі фічі будуть Glucose і BloodPressure.

<h3>Будуємо модельку<h3>

```python
log_reg_model = LogisticRegression()
log_reg_model.fit(x_train, y_train)
y_pred = log_reg_model.predict(x_test)
```

Порахуємо метрику Accuracy

```python
print(f"Accuracy = {log_reg_model.score(x_test, y_test)}")
```

Побудуємо ROC криву

```python
logit_roc_auc = metrics.roc_auc_score(y_test, y_pred)
fpr, tpr, threshols = metrics.roc_curve(y_test, y_pred)
plt.figure()
plt.plot(fpr, tpr, label="Logistic Regression (area = %0.2f)" % logit_roc_auc)
plt.plot([0,1], [0,1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Reciever operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

Порахуємо precision і recall

```python
precision = metrics.precision_score(y_test, y_pred)
recall = metrics.recall_score(y_test, y_pred)
print(f'precision equal {precision} і recall equal {recall}')
```

Отже, зараз наша Base-line моделька з точністю 78% може ідентифікувати наявність діабету і охоплює 63% усіх діабетиків

```python
cv_scores = cross_val_score(log_reg_model, x, y, cv=5)
```

```python
print(f'Середня accuracy модельки на кросс-валідції = {cv_scores.mean()}')
```

Спробуємо підібрати параметри по GridSearch

```python
std_slc = StandardScaler()
pca = decomposition.PCA()
logistic_Reg = LogisticRegression()

pipe = Pipeline(steps=[('std_slc', std_slc),
                           ('pca', pca),
                           ('logistic_Reg', logistic_Reg)])

n_components = list(range(1,x.shape[1]+1,1))
C = np.logspace(-4, 4, 50)
penalty = ['l1', 'l2']

parameters = dict(pca__n_components=n_components,
                      logistic_Reg__C=C,
                      logistic_Reg__penalty=penalty)

clf_GS = GridSearchCV(pipe, parameters)
clf_GS.fit(x_train, y_train)

print('Best Penalty:', clf_GS.best_estimator_.get_params()['logistic_Reg__penalty'])
print('Best C:', clf_GS.best_estimator_.get_params()['logistic_Reg__C'])
print('Best Number Of Components:', clf_GS.best_estimator_.get_params()['pca__n_components'])
print(); 
print(clf_GS.best_estimator_.get_params()['logistic_Reg'])
```

Навчаємо модельку на кращих гіперпараметрах

```python
std_slc = StandardScaler()
pca = decomposition.PCA()
logistic_Reg = LogisticRegression(C=0.18420699693267145, class_weight=None, dual=False,
          fit_intercept=True, intercept_scaling=1, max_iter=100,
        n_jobs=None, penalty='l2', random_state=None, 
            tol=0.0001, verbose=0, warm_start=False)

pipe = Pipeline(steps=[('std_slc', std_slc),
                           ('pca', pca),
                           ('logistic_Reg', logistic_Reg)])

pipe.fit(x_train, y_train)
y_pred_best = pipe.predict(x_test)
```

```python
# Подивимось на accuracy з кращими гіперпараметрами
print(f"Accuracy на кращих гіперпараметрах = {pipe.score(x_test, y_test)}")
```

```python
# Подивимось на accuracy з кращими гіперпараметрами на кросс валідації
cv_scores_best = cross_val_score(pipe, x, y, cv=5)
print(f'Середня accuracy модельки на кросс-валідції на кращих гіперпараметрах = {cv_scores_best.mean()}')
```

```python
# Порахуємо precision та recall з кращими гіперпараметрами 
precision_best = metrics.precision_score(y_test, y_pred_best)
recall_best = metrics.recall_score(y_test, y_pred_best)
print(f'Отже, на кращих гіперпараметрах precision equal {precision} і recall equal {recall}')
```

```python
# Ну і подивимось на ROC-AUC
logit_roc_auc = metrics.roc_auc_score(y_test, y_pred_best)
fpr, tpr, threshols = metrics.roc_curve(y_test, y_pred_best)
plt.figure()
plt.plot(fpr, tpr, label="Logistic Regression (area = %0.2f)" % logit_roc_auc)
plt.plot([0,1], [0,1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Reciever operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

В цілому підбір гіпер параметрів для модельки не покращив жодну з метрик, тож можна залишити base-line модельку.
