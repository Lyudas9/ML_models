## Створюємо X nf Y

```python
df = shuffle(df)
```

```python
x,y = df.drop(columns=['target','flavanoids']), df['target']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
def train_test_model(x_train, y_train, x_test, y_test,mod,
                    score = accuracy_score):
    mod.fit(x_train,y_train)
    preds_train = mod.predict(x_train)
    preds_test = mod.predict(x_test)
    cv_scores = cross_val_score(mod, x_train, y_train, cv=5)
    
    print('Train score')
    print(np.round(score(y_train,preds_train),2))
    print(f'Середня accuracy модельки на кросс-валідції train = {cv_scores.mean()}')
    
    print('Test score') 
    print(np.round(score(y_test,preds_test),2))
```

## Logistic Regression

```python
mod = LogisticRegression()
train_test_model(x_train, y_train, x_test, y_test,mod)
```

## SVC

```python
for kernel in ['linear','poly']:
    print(kernel)
    mod = SVC(kernel=kernel)
    train_test_model(x_train, y_train, x_test, y_test,mod)
    print()
```

```python
params = {'kernel':['linear','poly'],
         'degree': [2,3,4,5]}

svc = SVC()
svc.grid =GridSearchCV(svc,
                      params,
                      cv = 5,
                      verbose=4)
```

```python
svc.grid.fit(x_train,y_train)
```

```python
svc.grid.best_estimator_
```

```python
svc.grid.best_params_
```

```python
mod = SVC(degree=2, kernel='linear')
train_test_model(x_train, y_train, x_test, y_test, mod)
```

## Random Forest

```python
mod = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
train_test_model(x_train, y_train, x_test, y_test, mod)
```

```python
params = { 
    'n_estimators': [10, 50, 100],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}

rfc = RandomForestClassifier(random_state=42)
rfc.grid = GridSearchCV(rfc,
                      params,
                      cv = 5,
                      verbose=4)
```

```python
rfc.grid.fit(x_train,y_train)
```

```python
rfc.grid.best_estimator_
```

```python
rfc.grid.best_params_
```

```python
mod = RandomForestClassifier(random_state=42, max_features='auto', 
                             n_estimators= 50, max_depth=4, criterion='gini')
train_test_model(x_train, y_train, x_test, y_test, mod)
```

Отже виходить, що найбільшу якість має моделька по RandomForest Середня accuracy модельки на кросс-валідції train = 0.97, порахуємо ще додатково інші метрики.

```python
y_pred = mod.predict(x_test)
print(classification_report(y_test, y_pred))
```
